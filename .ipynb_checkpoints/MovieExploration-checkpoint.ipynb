## Import Dependencies

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## Read in CSV

```python
cleaned_df = pd.read_csv('data/movies_df_clean.csv')
cleaned_df
```

## Data Exploration!

```python
# Sorting values to see the top 5 by 'gross'

cleaned_df.sort_values(by=['Gross'], inplace=False, ascending=False).head(5)
```

```python
# Scatter plot budget vs gross

plt.scatter(x=cleaned_df['Budget'], y=cleaned_df['Gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot 

sns.regplot(x='Budget', y='Gross', data=cleaned_df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```
