## Import Dependencies

```python
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## Read in CSV

```python
cleaned_df = pd.read_csv('data/movies_df_clean.csv')
cleaned_df.head()
```

## Data Exploration!

### Hypothesis: If the budget of a film is high, then the gross-profit a movie generates will be high as well

```python
# Sorting values to see the top 5 by 'gross'

cleaned_df.sort_values(by=['Gross'], inplace=False, ascending=False).head()
```

```python
# Scatter plot budget vs gross

plt.scatter(x=cleaned_df['Budget'], y=cleaned_df['Gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')

plt.show()
```

```python
# Plot using seaborn to bring in a regression line

sns.regplot(x='Budget', y='Gross', data=cleaned_df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
# Identifying correlation in the data

cleaned_df.corr(method='pearson')
```

```python
# Visualizing as heatmap with Seaborn

correlation_matrix = cleaned_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matrix on Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Converting object columns into numeric values to include into correlation

df_numeric = cleaned_df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
        
df_numeric.head()
```

```python
# Creating updating heatmap with all features

correlation_matrix = df_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matrix on Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Grouping each combination of features

correlation_mat = df_numeric.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head(50)
```

```python
# Checking the the features with highest correlation greater than 0.5

highest_corr = sorted_pairs[(sorted_pairs) > 0.5]

highest_corr.head(20)
```

### We've identified two pairs of features having high correlation out of this dataset
#### Budget/Gross: 0.74
